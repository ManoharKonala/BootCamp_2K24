# **Understanding Seaborn** 

Imagine you're organizing a birthday party and want to make it as enjoyable and memorable as possible. To do that, you need to arrange everything neatly, from decorations and music to snacks and seating. Similarly, when you have a lot of data, Seaborn helps you arrange and visualize it in a neat and visually appealing way, making it easier to understand and analyze.


# **What is Seaborn?**

Seaborn is like a party planner for your data. It's a Python library built on top of Matplotlib that simplifies creating beautiful and informative charts and graphs. It helps you see patterns and trends in your data, making it easier to draw meaningful conclusions.

Imagine you are an artist with a blank canvas, and your data is the paint. Seaborn is your set of high-quality brushes that help you paint a beautiful and meaningful picture of your data.


# **Why Use Seaborn?**

**Prettier Plots:** Just like decorations make a party more attractive, Seaborn creates visually appealing graphs by default.

**More Control:** Seaborn is built on top of Matplotlib, so you can use all the Matplotlib features to customize your plots.

**Less Code:** Like a party planner saves you time, Seaborn lets you create complex plots with less code.

**Built-in Themes:** Seaborn comes with several themes, just like party themes, to style your plots.

**Integration with Pandas:** It works well with Pandas dataframes, making it easy to visualize data directly from these dataframes.

**1. DataFrames**
* *Analogy :* Think of a DataFrame like a spreadsheet or a table where each row is a record, and each column is a field.

# **Histogram**
* **Analogy:** Imagine counting how many people fall into different age groups.
### **Parameters:**

* data: Your DataFrame.

* x: Column for the x-axis.

* bins: Number of bins (age groups).

* hue: Optional, color bars by another variable.

* multiple: How to layer the bars (e.g., dodge, stack).

* kde: Whether to draw a kernel density estimate.

* element: Type of bars to draw (bars, step, poly).

* fill: Fill the area under the line.

* stat: Aggregate statistic to compute in each bin (count, frequency, density, probability).

* edgecolor: Edge color of bars.

* linewidth: Width of the bar edges.

```python
# Example DataFrame
data = sns.load_dataset('iris')

# Histogram
sns.histplot(data=data, x='sepal_length')
plt.show()
```

**Using KDE**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
scores = [78, 85, 92, 65, 88, 72, 90, 85, 80, 78, 82, 95, 68, 75, 84]

# Histogram
sns.histplot(scores, bins=5, kde=True, color='purple',binwidth=3, edgecolor='black', linewidth=1)
plt.title('Distribution of Test Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

**Using Multiple**

* Layer

```python
sns.histplot(data=data, x='sepal_length', hue='species', multiple='layer', kde=True)
plt.show()
```

**Element Parameter**


```python
sns.histplot(data=data, x="sepal_length", hue="species", element="step")
```

**Using stat**

statstr Aggregate statistic to compute in each bin.

count: show the number of observations in each bin

frequency: show the number of observations divided by the bin width

probability or proportion: normalize such that bar heights sum to 1

percent: normalize such that bar heights sum to 100

density: normalize such that the total area of the histogram equals 1

# **Bar Plot**
* Analogy: Imagine comparing the sales of different products.

### **Parameters:**

data: Your DataFrame.

x: Column for the x-axis.

y: Column for the y-axis.

hue: Optional, color bars by another variable.

ci: Confidence interval to draw around estimated values.

palette: Set color palette.

orient: Orientation of the plot (vertical or horizontal).

capsize: Size of the caps on error bars.

errcolor: Color of the error bars.

errwidth: Width of the error bars.

estimator: Function to estimate the value (e.g., mean, sum).

```python
# Example DataFrame
data = sns.load_dataset('titanic')

# Bar plot
sns.barplot(data=data, x='class', y='fare')
plt.show()
```

**Significance of black line in bars**

* The black line at the top of each bar is referred to as an "error bar." 
* The primary purpose of the error bar is to represent the variability or uncertainty about the measured value. 
* Visual Representation: The line typically extends from the top of the bar to both directions vertically. 
* The length of the line is determined based on the calculation (standard deviation, standard error, or confidence interval). 
* If the error bar is small, it suggests that the measurements are very consistent. A large error bar, conversely, suggests more variability and less certainty about the mean value.

**Imagine comparing the number of goals scored by different football teams.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
teams = ['Team A', 'Team B', 'Team C']
goals = [5, 3, 4]

# Bar plot
sns.barplot(x=teams, y=goals, palette='pastel', errorbar=None,hue=teams)
plt.title('Goals Scored by Teams')
plt.xlabel('Teams')
plt.ylabel('Goals')
plt.ylim(0, 6)
plt.show()
```

The errorbar API described was introduced in seaborn v0.12. In prior versions, the only options were to show a bootstrap confidence interval or a standard deviation, with the choice controlled by the ci parameter (i.e., ci=<size> or ci="sd").

```python
sns.barplot(data=data, x='class', y='fare', hue='sex', errorbar='sd', capsize=0.1, estimator = 'mean')
plt.show()

# sd = standard deviation
# se = standard error
# pi = percentile interval
# ci = confidence intervl

```

# **Rug plot**
* Analogy: Imagine marking the exact positions where cars are parked along a street.

* A rug plot is a simple way to visualize the distribution of data points along an axis. Each line represents a data point.

### **Parameters:**

* x: Data for the x-axis.
* height: Height of the rug lines.
* color: Color of the rug lines.
* lw: Line width of the rug lines.
* linestyle: Style of the rug lines ('-', '--', '-.', ':').

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
tips = sns.load_dataset('tips')
total_bill = tips['total_bill']

# Rug plot
sns.rugplot(x=total_bill, height=0.05, color='blue', lw=2, linestyle='--')
plt.title('Rug Plot of Total Bill')
plt.xlabel('Total Bill')
plt.show()
```

```python

# Example data
ages = [22, 25, 29, 35, 40, 42, 48, 50, 55, 60]

# Rug plot
sns.rugplot(ages, height=0.1, color='red', linewidth=2)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

# **Boxplot**
* Analogy: Imagine comparing the heights of students in different classes, with each box showing the range, median, and outliers.

* A boxplot provides a summary of a dataset's distribution, showing the median, quartiles, and potential outliers. It's useful for comparing distributions across different categories.

### **Parameters:**

* x: Categorical data for the x-axis.

* y: Numerical data for the y-axis.

* data: DataFrame containing the data.

* hue: Grouping variable to color the boxes.

* palette: Color palette for the boxes.

* linewidth: Width of the box lines.

* fliersize: Size of the outlier markers.

* orient: Orientation of the plot ('v' for vertical, 'h' for horizontal).

* width: Width of the boxes.

* notch: Whether to draw a notched box.

* showmeans: Whether to show the mean of the data.

* meanline: Whether to show the mean as a line.

* showfliers: Whether to show the outliers.

* whis: Length of the whiskers (1.5 by default).

```python
# Example data
tips = sns.load_dataset('tips')

# Boxplot
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='Set2', linewidth=2, fliersize=5)
plt.title('Boxplot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="time",showmeans=True,meanprops={"marker":"*","markeredgecolor":"b"},order=["Fri","Sat"],palette="plasma")
```

# **Violin Plot**
* Analogy: Imagine comparing the distribution of exam scores for different classes, with each "violin" showing the density of the scores.

* A violin plot combines aspects of a boxplot and a density plot. It shows the distribution of the data across different categories, with the width of each "violin" representing the density of the data.

### **Parameters:**

* x: Categorical data for the x-axis.

* y: Numerical data for the y-axis.

* data: DataFrame containing the data.

* hue: Grouping variable to color the violins.

* palette: Color palette for the violins.

* split: Whether to split the violins for different levels of the hue variable.

* density_norm: Method for scaling the width of each violin ('area', 'count', 'width').

* inner: Representation of the datapoints in the violin interior ('box', 'quartile', 'point', 'stick').

* bw_method: Bandwidth for the kernel density estimate.


```python
# Example data
tips = sns.load_dataset('tips')

# Violin plot
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True, palette='Set1', density_norm='width', inner='quartile', bw_method=0.2)
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Example data
classes = ['Class A', 'Class B', 'Class C']
scores = [78, 82, 85, 88, 90, 72, 75, 80, 82, 85, 88, 90, 60, 65, 70, 75, 80, 85]

# Violin plot
sns.violinplot(x=classes * 6, y=scores, inner='quartile', density_norm='width', bw_method=0.5)
plt.title('Test Scores Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```
