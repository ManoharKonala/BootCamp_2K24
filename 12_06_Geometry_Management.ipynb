# Place
it is very useful when you want more control over the exact location of your widgets.

The place layout manager allows for you to have more absolute control about the arrangement of your widgets. With place, you can specify the size of the widget, as well as the exact x and y coordinates to arrange it within the parent window.

It allows you explicitly set the position and size of a window, either in absolute terms, or relative to another window

Usage: Absolute positioning of widgets.

Control: Complete control over widget placement.

Best For: Custom layouts requiring precise widget placement.

Parameters:
* in_ -- specifies the master window for the widget
* x, y -- specifies the specific x and y values of the widget in the parent window
* relx, rely -- horizontal and vertical offset relative to the size of the parent widget, values between 0.0 and 1.0
* relwidth, relheight -- set height and width of widget relative to the size of the parent widget, values between 0.0 and 1.0
* anchor -- where the widget is placed in the parent widget, specified by 'n', 's', 'e', 'w', or some combination of them. Default is "CENTER".

```python
from tkinter import *

root = Tk()
root.geometry("200x200")

L = Label(root, text="place manager")
L.place(anchor="se", x=200, y=200)  # Positioning the label at the bottom right corner

button1 = Button(root, text="Button1")
button1.place(relx=0.5, rely=0.5, anchor="center", width=100, height=50)
# Positioning button1 at the center of the window with specified width and height

button2 = Button(button1, text="Button2")
button2.place(relx=0.5, rely=0.5, anchor="center")
# Positioning button2 at the center of button1

root.mainloop()

```

# Grid

The grid method organizes widgets in a 2-dimensional table-like structure (think of an Excel spreadsheet), where each cell can hold a widget. You don't have to specify the size of the widgets in your window as the manager already handles that for you.

The top left corner has values row=0 and column=0, and increase by 1 as you move right or down in the grid. If you want more control over arranging widgets or resizing them, it is also possible to span across multiple rows



Usage: Organizing widgets in a table-like structure.

Control: More control over widget alignment and spacing.

Best For: Complex forms and multi-widget layouts.

Notice below how each widget can be positioned within "cells" by specifying their row and column values.

image.png

 Don't mix grid() and pack() methods. The application will freeze trying to figure out how to make these two managers work together.

Parameters:
* row, column -- the row and column values for the location of the widget. Both start from 0.
* padx, pady -- Adds horizontal and vertical padding around the widget...
* ipadx, ipady -- Adds internal horizontal and vertical padding within the widget.
* sticky -- specify a value of 's', 'n', 'e', 'w', or a combination of them, e.g. 'nw', 'ne', 'sw', or 'se'. The parameter tells which side of the 'cell' the widget will 'stick' to .  

If you use 'w'+'e'+'n'+'s', then the widget will fill up the 'cell'. Default is to center the widget within the 'cell'.

```python
import tkinter as tk

root = tk.Tk()
root.geometry("250x250")

root.title("GRID")

tk.Label(root, text="Name").grid(row=0, column=0, padx=20, pady=0,sticky="NE")
tk.Entry(root).grid(row=0, column=1, padx=5, pady=5, sticky="W")

tk.Label(root, text="Email").grid(row=1, column=0, padx=20, pady=0, sticky="E")
tk.Entry(root).grid(row=1, column=1, padx=5, pady=5, sticky="W")

btn2 = tk.Button(root, text="Bottom Right")
btn2.grid(row = 2, column = 1,sticky='se')

btn3 = tk.Button(root, text="Fill Horizontally")
btn3.grid(row=3, column=0, sticky='ew')
```

# Pack

The pack() method turns each individual widget into a block. Each widget has its own size and the pack manager fits them all together just like you would do with real blocks.



Widgets are organized into blocks and placed into the parent window depending upon specified parameters. But unlike grid, you don't need to specify the row and column values. Widgets can be placed vertically or horizontally depending upon the values for the side parameter. For more complex layouts, combine pack with Frame widgets.

Usage: Simple, linear stacking of widgets.

Control: Limited control over precise placement.

Best For: Quick and simple layouts.

Parameters:
* side: Specifies which side of the parent widget the child widget should be packed against. Accepts tk.TOP, tk.BOTTOM, tk.LEFT, or tk.RIGHT. (default is 'top').

* fill: which directions you want the widget to fill in the parent window, can choose 'x', 'y' directions, or 'both'.. Accepts tk.NONE, tk.X, tk.Y, or tk.BOTH.

* expand: When set to True, the widget expands to fill any extra space in the geometry master. Default is "False".

* anchor: Determines which part of the widget is anchored to the position defined by side. Accepts values like tk.N, tk.S, tk.E, tk.W, tk.NE, tk.NW, tk.SE, tk.SW, or tk.CENTER.
* padx: Adds horizontal padding around the widget.
* pady: Adds vertical padding around the widget.
* padx: Adds internal horizontal padding within the widget.
* ipady: Adds internal vertical padding within the widget.

```python
import tkinter as tk

root = tk.Tk()
root.title("Pack Geometry Manager Example")

# Create frames to organize the layout
top_frame = tk.Frame(root, bg="lightblue")
bottom_frame = tk.Frame(root, bg="lightgreen")

top_frame.pack(side=tk.TOP, fill=tk.BOTH, expand=True, padx=10, pady=10)
bottom_frame.pack(side=tk.BOTTOM, fill=tk.BOTH, expand=True, padx=10, pady=10)

# Add labels to the top frame with different pack options
label1 = tk.Label(top_frame, text="Top Left", bg="red")
label2 = tk.Label(top_frame, text="Top Center", bg="yellow")
label3 = tk.Label(top_frame, text="Top Right", bg="blue")

label1.pack(side=tk.LEFT, fill=tk.Y, expand=True, padx=5, pady=5, ipadx=10, ipady=10, anchor=tk.N)
label2.pack(side=tk.TOP, fill=tk.BOTH, expand=True, padx=5, pady=5, ipadx=10, ipady=10, anchor=tk.CENTER)
label3.pack(side=tk.RIGHT, fill=tk.Y, expand=True, padx=5, pady=5, ipadx=10, ipady=10, anchor=tk.S)

# Add buttons to the bottom frame with different pack options
button1 = tk.Button(bottom_frame, text="Bottom Left", bg="orange")
button2 = tk.Button(bottom_frame, text="Bottom Center", bg="purple")
button3 = tk.Button(bottom_frame, text="Bottom Right", bg="pink")

button1.pack(side=tk.LEFT, fill=tk.X, expand=True, padx=5, pady=5, ipadx=10, ipady=10, anchor=tk.W)
button2.pack(side=tk.BOTTOM, fill=tk.BOTH, expand=True, padx=5, pady=5, ipadx=10, ipady=10, anchor=tk.S)
button3.pack(side=tk.RIGHT, fill=tk.X, expand=True, padx=5, pady=5, ipadx=10, ipady=10, anchor=tk.E)

root.mainloop()
```
