# What is Matplotlib?
Imagine you're an artist with a blank canvas, paintbrushes, and a palette of colors. Matplotlib is like your artist toolkit in the world of Python programming. It allows you to create beautiful and informative visualizations (like charts and graphs) from data.


# Components of Matplotlib

**Canvas (Figure):**

* Analogy: The canvas you paint on.

* Real-world example: When you start painting, you need a blank piece of paper or a canvas. In Matplotlib, this blank canvas is called a "Figure."

* Python: plt.figure()

**Axes (Plot Area):**

* Analogy: The specific area on the canvas where you paint.

* Real-world example: On your canvas, you might have different sections for different parts of your painting. In Matplotlib, these sections are called "Axes."

* Python: fig.add_subplot(1,1,1) or simply plt.subplot()

**Brushes and Colors (Plots):**

* Analogy: The brushes and colors you use to paint.

* Real-world example: To create a painting, you use brushes of different sizes and colors of paint. In Matplotlib, these brushes and colors are like different types of plots (line plots, bar charts, scatter plots, etc.).

* Python: plt.plot(), plt.bar(), plt.scatter(), etc.

**Creating a Simple Plot**

* x and y: Lists of values representing the data points.

* plt.plot(x, y): Creates a line plot with x on the horizontal axis and y on the vertical axis.

* plt.xlabel('X-axis'): Adds a label to the x-axis.

* plt.ylabel('Y-axis'): Adds a label to the y-axis.

* plt.title('Simple Line Plot'): Adds a title to the plot.

* plt.show(): Displays the plot.


```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Creating a plot
plt.plot(x, y)

# Adding labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Displaying the plot
plt.show()
```

**1.Set up your canvas:**

* Analogy: Get your blank canvas ready.

**2.Define your plot area:**

* Analogy: Decide where on the canvas you'll draw your line.

**3.Draw your line:**

* Analogy: Choose your brush and color, then draw a line.

* Real-world example: Imagine you're drawing the path of a car's journey over time on your canvas.

**4.Add finishing touches:**

* Analogy: Add details like labels and a title to your painting.

**5.Display your masterpiece:**

* Analogy: Show your finished painting to others.


```python
import matplotlib.pyplot as plt

# Set up your canvas
fig = plt.figure()

# Define your plot area
ax = fig.add_subplot(1, 1, 1)

# Draw your line
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Add finishing touches
ax.set_title('Car Journey Over Time')
ax.set_xlabel('Time (hours)')
ax.set_ylabel('Distance (miles)')

# Display your masterpiece
plt.show()
```

**Line Plot**

* Used to display data points connected by straight lines. Good for showing trends over time.


* x, y: Data points.

* color: Sets the color of the line.

* marker: Specifies the marker style for the data points (e.g., 'o' for circle).

* label='Line': Adds a label to the line, which will be used in the legend.

* plt.legend(): Displays the legend.


```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, label='Prime Numbers', color='blue', marker='o')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot Example')
plt.legend()
plt.show()
```

**Scatter Plot**
* Displays individual data points. Great for showing relationships between two variables.
* x, y: Data points.
* color: Sets the color of the points.
* marker: Specifies the marker style (e.g., 'x' for cross).

* plt.scatter(x, y, color='red'): Creates a scatter plot with red markers.

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78, 77, 85, 86]

plt.scatter(x, y, color='red', marker='x')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot Example')
plt.show()
```

**Bar Plot**
* Useful for comparing different categories.
* x, y: Data points.
* color: Sets the color of the bars.
* width: Sets the width of the bars
* plt.bar(x, y, color='blue'): Creates a bar plot with blue bars.

```python
x = ['A', 'B', 'C', 'D']
y = [3, 7, 5, 12]

plt.bar(x, y, color='green', width=0.5)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

**Histogram**
* Shows the distribution of a dataset.

* data: Data points.
* bins: Number of bins (bars) in the histogram.
* color: Sets the color of the bars.
* edgecolor: Sets the color of the bar edges

* plt.hist(data, bins=5, color='green'): Creates a histogram with 5 bins and green bars.

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

plt.hist(data, bins=4, color='purple', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

**Pie Chart**

* sizes: Data points (sizes of each slice).
* explode: Moves slices away from the center.
* labels: Labels for each slice.
* colors: Colors for each slice.
* autopct: Displays percentage value.
* shadow: Adds a shadow.
* startangle: Rotates the start of the pie chart.

* plt.pie(sizes, labels=labels, autopct='%1.1f%%'): Creates a pie chart with percentages displayed.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Pie Chart Example')
plt.show()
```

# Subplots


**Creating Multiple Plots**

This example shows how to create multiple plots in the same figure window. You can create multiple plots in the same figure window by calling the plot function multiple times. Each call to plot creates a new plot.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, y, color='blue')
ax1.set_title('Plot 1')

ax2.plot(y, x, color='green')
ax2.set_title('Plot 2')

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[4,5,6,3,2,3]
plt.plot(x,y,linestyle='-.',marker='*')
#linestyle is used to change the formatting of line so that we can easily identify the lije

x1=[2,4,6,8,10]
y1=[1,2,3,4,5]
plt.plot(x1,y1,color="red",marker="o")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Multiple plot")
plt.show()
# If we write plt.show() after each graph then it will draw a separate graph after each graph 
# if we write plt.show() after two plots then it will combine both the graphs
```

**Adjusting Layout**


```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout()
plt.show()
```
